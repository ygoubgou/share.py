```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1) Dictionnaire sets d'individus par vague
wave_ids = df_1_clean.groupby('wave')['mergeid'].apply(set)

# 2) Initialiser la matrice 9x9
waves = sorted(wave_ids.index)
matrix = np.zeros((len(waves), len(waves)), dtype=int)

# 3) Remplir la matrice avec les tailles des intersections
for i, w1 in enumerate(waves):
    for j, w2 in enumerate(waves):
        matrix[i, j] = len(wave_ids[w1].intersection(wave_ids[w2]))

# 4) Afficher la matrice avec seaborn heatmap
plt.figure(figsize=(8,6))
sns.heatmap(matrix, xticklabels=waves, yticklabels=waves, annot=True, fmt='d', cmap='Blues')
plt.title('Nombre d\'individus communs entre vagues')
plt.xlabel('Vague')
plt.ylabel('Vague')
plt.show()
```

```python
df = df_ap
# 1 = traitement (veuf), 0 = contrôle (non veuf)
df['traitement'] = df['changement_statut']
df = df.dropna()
# Garder mergeid à part
mergeid = df["mergeid"]
fin_cat = df["co007_"]
```

```python
from sklearn.preprocessing import OrdinalEncoder

cat_cols = df.select_dtypes(include='object').columns
encoder = OrdinalEncoder()
df[cat_cols] = encoder.fit_transform(df[cat_cols])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler



# On enlève la variable cible `casp` et le statut pour le matching
X = df.drop(columns=["casp", "changement_statut", "traitement", "mergeid","bmi2"])
y = df["traitement"]

# Standardisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Régression logistique pour les scores
logit = LogisticRegression()
logit.fit(X_scaled, y)

# Ajout des scores dans le DataFrame
df["propensity_score"] = logit.predict_proba(X_scaled)[:, 1]

# Réajout explicite de mergeid pour ne pas le perdre
df["mergeid"] = mergeid
df["co007_"] = fin_cat 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(data=df, x="propensity_score", hue="traitement", common_norm=False)
plt.title("Distribution des scores de propension")
plt.xlabel("Propensity Score")
plt.ylabel("Densité")
plt.legend(title="Groupe", labels=["Contrôle", "Traitement"])
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors

# Séparer les groupes
traitement = df[df['traitement'] == 1].copy()
controle = df[df['traitement'] == 0].copy()

# Apprentissage sur le groupe contrôle
nn = NearestNeighbors(n_neighbors=1)
nn.fit(controle[["propensity_score"]])

# Recherche des plus proches voisins
distances, indices = nn.kneighbors(traitement[["propensity_score"]])

# Extraction des correspondances
controle_matched = controle.iloc[indices.flatten()].copy()
controle_matched["matched_mergeid"] = traitement["mergeid"].values
traitement["matched_mergeid"] = traitement["mergeid"].values  # Pour alignement

# Fusion
df_matched = pd.concat([traitement, controle_matched], ignore_index=True)

# On garde uniquement les colonnes nécessaires
df_matched[["mergeid", "matched_mergeid", "traitement", "propensity_score"]]
```

```python
#correlations
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from statsmodels.stats.multitest import multipletests

# Colonnes sélectionnées
cols = ["eduyears_mod", "age", "thinc_m", "bmi","hhsize","eurod","casp","maxgrip", "hc002_mod","chronic_mod"]
df = donnees_R[cols].dropna()

# Initialisation
n = len(cols)
cor_matrix = np.zeros((n, n))
pval_matrix = np.zeros((n, n))
adjusted_pvals = np.zeros((n, n))

# Corrélations et p-values
for i in range(n):
    for j in range(n):
        if i <= j:
            r, p = pearsonr(df.iloc[:, i], df.iloc[:, j])
            cor_matrix[i, j] = cor_matrix[j, i] = r
            pval_matrix[i, j] = pval_matrix[j, i] = p

# Ajustement des p-values (méthode FDR)
pvals_flat = pval_matrix[np.triu_indices(n, k=1)]
_, adj_pvals_flat, _, _ = multipletests(pvals_flat, method='fdr_bh')
triu_indices = np.triu_indices(n, k=1)
for idx, (i, j) in enumerate(zip(*triu_indices)):
    adjusted_pvals[i, j] = adjusted_pvals[j, i] = adj_pvals_flat[idx]

# Conversion en DataFrames
cor_df = pd.DataFrame(cor_matrix, index=cols, columns=cols)
adj_pval_df = pd.DataFrame(adjusted_pvals, index=cols, columns=cols)

# ==== 📊 Heatmap Corrélation ====
plt.figure(figsize=(8, 6))
sns.heatmap(cor_df, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Corrélation de Pearson")
plt.show()

# ==== 📊 Heatmap p-values ajustées ====
plt.figure(figsize=(8, 6))
sns.heatmap(adj_pval_df, annot=True, fmt=".3f", cmap="YlGnBu", vmin=0, vmax=1)
plt.title("p-values ajustées (méthode FDR)")
plt.show()

# === Figure partagée ===
fig, axes = plt.subplots(1, 2, figsize=(16, 6))  # 1 ligne, 2 colonnes

# Heatmap de corrélation
sns.heatmap(cor_df, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1, ax=axes[0])
axes[0].set_title("Corrélation de Pearson")

# Heatmap des p-values ajustées
sns.heatmap(adj_pval_df, annot=True, fmt=".3f", cmap="YlGnBu", vmin=0, vmax=1, ax=axes[1])
axes[1].set_title("p-values ajustées ")

plt.tight_layout()
plt.show()

```
